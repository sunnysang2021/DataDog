```python
import warnings
warnings.simplefilter('ignore')

import datetime
import numpy as np
import pandas as pd
import seaborn as sns
import graphviz
import matplotlib.pyplot as plt
from matplotlib import rcParams
from sklearn.metrics import auc, roc_curve, classification_report

%matplotlib inline
```

# Objective

# Functions

```python
def read_csv_file(file_name):
    df_name = pd.read_csv(file_name)
    return df_name
```

```python
def convert_timestamp_to_date(df_name, timestamp_col_name):
    return pd.to_datetime(df_name[timestamp_col_name]).dt.strftime('%Y-%m-%d')
```

```python
def pageview_cnt(df_name, groupby_list):
    pageview_cnt_by_col = df_name.groupby(groupby_list).pageview_id.count().reset_index().rename(columns={"pageview_id":"cnt"})
    return pageview_cnt_by_col
```

```python
def create_pageview_cnt_line_chart(title, df_name, x_col, y_col, hue_col):
    rcParams['figure.figsize'] = 15.7,12.27
    if hue_col == 'NaN':
        sns.lineplot(x = pd.to_datetime(df_name[x_col])
                 , y = df_name[y_col]
                 , palette = 'bright'
                )
    else:
        sns.lineplot(x = pd.to_datetime(df_name[x_col])
                 , y = df_name[y_col]
                 , hue = df_name[hue_col]
                 , palette = 'bright'
                )
    plt.xticks(rotation=30)
    plt.title(title)
```

```python
def join_table(left_df_name, right_df_name, left_index, right_index, lsuffix_name, rsuffix_name):
    left = left_df_name.set_index(left_index)
    right = right_df_name.set_index(right_index)
    if lsuffix_name == 'NaN' or rsuffix_name == 'NaN':
        combined = left.join(right).reset_index()
    else:
        combined = left.join(right, lsuffix = lsuffix_name, rsuffix = rsuffix_name).reset_index()
    return combined.rename(columns={"index":left_index})
```

```python
def filter_out_top_N_variable_list(df_name, variable, n):
    top10_variable = df_name[variable].value_counts().reset_index().rename(columns={"index":variable, variable:"cnt"}).nlargest(n,'cnt')
    top10_variable_list = top10_variable[variable].tolist()
    return top10_variable_list
```

# Analysis

```python
# read all 4 csv files
fact_app_pageview = read_csv_file('fact_app_pageview.csv')
dim_app_page = read_csv_file('dim_app_page.csv')
dim_customer_company = read_csv_file('dim_customer_company.csv')
dim_user = read_csv_file('dim_user.csv')
```

```python
fact_app_pageview['pageview_date'] = convert_timestamp_to_date(fact_app_pageview, 'pageview_timestamp')
```

```python
total_pageview_cnt = pageview_cnt(fact_app_pageview,['pageview_date'])
create_pageview_cnt_line_chart('Daily Pageview Count',total_pageview_cnt, 'pageview_date','cnt','NaN')
```

```python
# check the trend of daily active user over the time
daily_active_users = fact_app_pageview.groupby('pageview_date').apply(lambda x: len(x.datadog_user_id.unique())).reset_index()
daily_active_users = daily_active_users.rename(columns={0: 'cnt'})
create_pageview_cnt_line_chart("Daily Active Users", daily_active_users, 'pageview_date', 'cnt', 'NaN')
```

```python
# check the trend of daily active customer company over the time
daily_active_company = fact_app_pageview.groupby('pageview_date').apply(lambda x: len(x.company_id.unique())).reset_index()
daily_active_company = daily_active_company.rename(columns={0: 'cnt'})
create_pageview_cnt_line_chart("Daily Active Customer Company", daily_active_company, 'pageview_date', 'cnt', 'NaN')
```

```python
# look into geo-country
pageview_cnt_by_geo = pageview_cnt(fact_app_pageview,['pageview_date','geo_country'])

top10_geo_lsit = filter_out_top_N_variable_list(fact_app_pageview, 'geo_country', 10)
```

```python
pageview_cnt_by_top10_geo = pageview_cnt_by_geo[pageview_cnt_by_geo.geo_country.isin(top10_geo_lsit)]
create_pageview_cnt_line_chart('Daily Pageview Count in Top 10 Country',pageview_cnt_by_top10_geo, 'pageview_date','cnt','geo_country')
```

```python
# join fact_app_pageview with dim_app_page
main_table = join_table(fact_app_pageview, dim_app_page, 'page_id', 'id', 'NaN', 'NaN')
main_table.head()
```

```python
# join main table with dim_customer_company
main_table = join_table(main_table, dim_customer_company, 'company_id', 'id', '_pageview', '_company')
main_table.head()
```

```python
# join main table with dim_user
main_table = join_table(main_table, dim_user, 'datadog_user_id', 'id','_','_user')
main_table.rename(columns={'name_':'company_name','name_user':'user_name'})
main_table.head()
```

```python
pageview_cnt_by_category = pageview_cnt(main_table,['pageview_date','category'])
create_pageview_cnt_line_chart('Daily Pageview Count by Category',pageview_cnt_by_category, 'pageview_date','cnt','category')
```

```python
# filter out rows in Dashboards
dashboards = main_table[main_table.category == 'dashboards']
```

```python
pageview_cnt_by_billing_plan_in_Dashboards = pageview_cnt(dashboards,['pageview_date','billing_plan'])
create_pageview_cnt_line_chart('Daily Pageview Count by Billing Plan in Dashboards',pageview_cnt_by_billing_plan_in_Dashboards, 'pageview_date','cnt','billing_plan')
```

# Break Down into Billing Plan

```python
# look into Dogfood
dogfood = dashboards[dashboards.billing_plan == 'dogfood']
pageview_cnt_in_Dogfood_in_Dashboards = pageview_cnt(dogfood,['pageview_date'])
create_pageview_cnt_line_chart('Daily Pageview Count in DogFood in Dashboards',pageview_cnt_in_Dogfood_in_Dashboards, 'pageview_date','cnt','NaN')
```

```python
# look into enterprise
enterprise = dashboards[dashboards.billing_plan == 'enterprise']

# filter out Top 10 Company in Enterprise
top10_comp_in_enterprise_list = filter_out_top_N_variable_list(enterprise, 'company_id', 10)
pageview_cnt_by_top10_comp_in_enterprise = enterprise[enterprise.company_id.isin(top10_comp_in_enterprise_list)]
pageview_cnt_in_Enterpise_in_Dashboards_at_comp_level = pageview_cnt(pageview_cnt_by_top10_comp_in_enterprise,['pageview_date','company_id'])

create_pageview_cnt_line_chart('Daily Pageview Count in Enterprise in Dashboards at Company Level',pageview_cnt_in_Enterpise_in_Dashboards_at_comp_level, 'pageview_date','cnt','company_id')
```

```python
# look into pro
pro = dashboards[dashboards.billing_plan == 'pro']

# filter out Top 10 Company in Pro
top10_comp_in_pro_list = filter_out_top_N_variable_list(pro, 'company_id', 10)
pageview_cnt_by_top10_comp_in_pro = pro[pro.company_id.isin(top10_comp_in_pro_list)]
pageview_cnt_in_pro_in_Dashboards_at_comp_level = pageview_cnt(pageview_cnt_by_top10_comp_in_pro,['pageview_date','company_id'])

create_pageview_cnt_line_chart('Daily Pageview Count in Pro in Dashboards at Company Level',pageview_cnt_in_pro_in_Dashboards_at_comp_level, 'pageview_date','cnt','company_id')
```
