# Join fact_app_pageview with dim_app_page

```python
dim_app_page = pd.read_csv('dim_app_page.csv')
dim_app_page.head()
```

```python
left = fact_app_pageview.set_index('page_id')
right = dim_app_page.set_index('id')

combined = left.join(right).reset_index()
combined.head()
```

```python
category_cnt = combined.groupby(['category','pageview_date']).pageview_id.count().reset_index().rename(columns={"pageview_id":"cnt"})
category_cnt
```

```python
rcParams['figure.figsize'] = 15.7,12.27
sns.lineplot(x = pd.to_datetime(category_cnt["pageview_date"])
             , y = category_cnt["cnt"]
             , hue = category_cnt["category"]
             , palette = 'bright'
            )
plt.xticks(rotation=30)
plt.title("Daily Pageview Count by Cateogry")
```

```python
dashboards = combined[combined.category == 'dashboards']
dashboards
```

```python
# define a function to verify the unique values in each categorical column
def check_unique(df_name, col_name):
    print('-- ',col_name, ': ', df_name[col_name].unique())
```

```python
# check outliers for price
def plotBoxplot(table_name, col_name):
    sns.boxplot(table_name[col_name])
    plt.title("Box Plot before outlier removing in '" + str(col_name) + "'")
    plt.show()
```

```python
groupby_company_id = combined3.groupby(['company_id']).pageview_id.count().sort_values(ascending=False).reset_index().rename(columns={"pageview_id": "cnt"})
```

```python
plotBoxplot(groupby_company_id, 'cnt')
```

```python
groupby_company_id
```

```python
top18_company_list = groupby_company_id[groupby_company_id.cnt >= 2000]
top18_company_list = top18_company_list.company_id.tolist()
top18_company_list
```
